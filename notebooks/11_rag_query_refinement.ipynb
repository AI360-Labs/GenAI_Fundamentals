## âš™ï¸ Section 1: Setup and API Configuration


```python
import os
from google import genai
from google.colab import userdata
```

```python
# --- 1. SETUP (Requires Gemini API Key) ---
# NOTE: You can get a key from Google AI Studio.

api_key = userdata.get('GOOGLE_API_KEY_1')

# Initialize the Gemini client
client = genai.Client(api_key=api_key)

gemini_model = 'gemini-2.5-flash'
```

```python
def rewrite_query(user_query: str) -> str:
    """
    Uses the Gemini LLM to rewrite a vague user query into a precise search query.
    """
    # The prompt instructing the LLM on the rewriting task
    SYSTEM_PROMPT = (
        "You are an expert search query optimizer. Your job is to take the "
        "user's short or ambiguous question and rewrite it into a single, "
        "detailed, high-quality search query that will maximize relevant "
        "results from a technical knowledge base. Output only the new query."
    )

    # 2. API Call to the Gemini Model
    try:
        response = client.models.generate_content(
            model=gemini_model,
            contents=[
                {"role": "user", "parts": [{"text": SYSTEM_PROMPT + "\n\nOriginal Query: " + user_query}]},
            ],
            config={"temperature": 0.0} # Keep the output deterministic
        )
        # 3. Extract and Clean the Rewritten Query
        return response.text.strip()
    except Exception as e:
        return f"Error during API call: {e}"
```

```python
# --- 4. EXECUTION ---
vague_user_input = "energy consumption" # Try other questions!

rewritten_query = rewrite_query(vague_user_input)

print("="*50)
print("Query Refinement using Gemini")
print("="*50)
print(f"**Vague User Input:**\n{vague_user_input}")
print("-" * 25)
print(f"**Gemini Rewritten Query:**\n{rewritten_query}")
print("-" * 25)
```
